```python
import cv2
import numpy as np

def video_to_numpy(video_path):
    # 비디오 캡처 객체 생성
    cap = cv2.VideoCapture(video_path)
    
    if not cap.isOpened():
        print(f"Error: Unable to open video file {video_path}")
        return None
    
    # 비디오의 프레임 수, 너비, 높이, 채널 수 가져오기
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)

    print(f"Frame count: {frame_count}, Width: {width}, Height: {height}, FPS: {fps}")

    # NumPy 배열 초기화
    video_array = np.empty((frame_count, height, width, 3), np.dtype('uint8'))

    # 비디오의 각 프레임을 NumPy 배열에 저장
    for i in range(frame_count):
        ret, frame = cap.read()
        if not ret:
            print(f"Error: Unable to read frame {i}")
            break
        video_array[i] = frame

    # 비디오 캡처 객체 해제
    cap.release()
    
    return video_array

# 예제 사용법
video_path = 'rendering.mp4'
video_array = video_to_numpy(video_path)
print(video_array.shape)  # (프레임 수, 높이, 너비, 채널 수)
```

```python
def numpy_to_video(video_array, output_path, fps):
    # 비디오 배열의 차원 확인
    frame_count, height, width, channels = video_array.shape

    # 비디오 라이터 객체 생성
    fourcc = cv2.VideoWriter_fourcc(*'H264')  # 또는 'XVID', 'MJPG' 등으로 변경 가능
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    # 각 프레임을 비디오 파일에 쓰기
    for i in range(frame_count):
        frame = video_array[i]
        out.write(frame)

    # 비디오 라이터 객체 해제
    out.release()

# 예제 사용법
# 랜덤 데이터로 예제 영상 생성
frame_count, height, width, channels = video_array.shape
fps = 30.301950654671774

output_path = 'rendering_real.mp4'

numpy_to_video(video_array, output_path, fps)
print(f"Video saved to {output_path}")
```
